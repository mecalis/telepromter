A nagyon sok, ~200+ db, országkód miatt azokat az országokat veszem be az elemzésbe, akik "értékelhető" felhasználói számmal képviseltetik magukat. Ez az alsó limit a "limit" változóval szabályozható.
A végeredményt oszlopdiabrammon ábrázolom.

```python
#Szűrjük darabszámra, min ennyi user
limit = 40
country_df_group_to_plot = country_df_group.filter(pl.col("pcs") > limit).to_pandas()
country_df_group_to_plot['country_code_ordered'] = pd.Categorical(country_df_group_to_plot['country_code'], categories=country_df_group_to_plot['country_code'].unique(), ordered=True)

country_df_group_plot = (
    ggplot(country_df_group_to_plot, aes(x='country_code_ordered', y='pcs', fill='event'))  # Ábrázolás
    + geom_bar(stat='identity', position='stack')  # Stacked bar plot
    + coord_flip()  # Vízszintes tengelyek
    + theme(axis_text_x=element_text(rotation=45, hjust=1), figure_size=(8, 6))  # X tengely címek forgatása
    + labs(title="Országok szerinti cancel vs. convert", x="Országok", y="Felhasználók száma")
    + scale_fill_manual(values={"trial_converted_event": "green", "trial_cancelled_event": "red"})
    
)

display(country_df_group_plot)
```

```python
# Az események szétválasztása, trial_cancelled_event / trial_converted_event hányados számítása
df_pivoted = (
    country_df_group.pivot(
        values="pcs", 
        index="country_name", 
        on="event", 
        aggregate_function="first"
    )
)

df_pivoted = df_pivoted.drop_nulls()
df_pivoted = df_pivoted.with_columns(
    (((pl.col("trial_converted_event") / (pl.col("trial_cancelled_event") + pl.col("trial_converted_event"))))*100).alias("cancelled_to_converted_ratio (%)")
).filter(pl.col("trial_converted_event") > limit).sort(pl.col("cancelled_to_converted_ratio (%)"), descending = True)
#df_pivoted.to_pandas()
```

## Országonkénti GDP arányos elemzés. 
Arra a kérdésre keresem a választ, hogy az országonkénti eltérő konverziós arány összefüggésben van-e az ország GDP-jével.

```python
def generate_circle(x, y, radius, ratio) -> pd.DataFrame:
    '''
    Kör rajzoló segédfüggvény, amivel egy ggplot ábrán karikázni lehet.
    x : float
        Kör közepének koordinátája a vízszintes tengelyen.
    y : float
        Kör közepének koordinátája a függőleges tengelyen.
    radius : float
        A kör sugara.
    ratio : float
        A kör főtengelyinek aránya, amivel az ellipszis torzítást vezérelhetjük. 
        Elsősorban az x és y változók léptékeinek különbsége miatt van rá szükség.
    '''
    theta = np.linspace(0, 2*np.pi, 100)
    circle_x = (x + radius * ratio *np.cos(theta)) 
    circle_y = (y + radius * np.sin(theta))
    return pd.DataFrame({"x": circle_x, "y": circle_y})

circle_df_poor = generate_circle(x = 19000, y=12, radius=15, ratio=800)
circle_df_rich = generate_circle(x = 79000, y=26, radius=16, ratio=2000)
gdp_plot = (
    ggplot(df_pivoted3, aes(x="GDP", y="cancelled_to_converted_ratio (%)")) +
    geom_point() +
    geom_smooth() +
    theme_minimal() +
    theme(axis_text_x=element_text(angle=45), figure_size=(8, 8), plot_caption=element_text(size=10, color="gray", ha="left")) +
    geom_path(data=circle_df_poor, mapping=aes(x="x", y="y"), color="red", size=1, alpha =0.2) +
    geom_path(data=circle_df_rich, mapping=aes(x="x", y="y"), color="blue", size=1, alpha =0.2) +
    geom_text(aes(label="country_name"), va="bottom", ha="center", size=8, angle=0) +
    labs(title="Országonkénti konverziós arány az egy főre eső \n        jövedelem függvényében", 
         x="Egy főre jutó GDP, vásárlóerő-paritás alapján [USD]", 
         y="Konverziós arány [%]") + 
    annotate("text", x=30000, y=2, label="← Fejlődő országok, alacsonyabb bevétellel", ha="left") +
    annotate("text", x=27000, y=28, label="Jóléti államok →", ha="left") +
    labs(caption="Forrás: https://en.wikipedia.org/wiki/List_of_countries_by_GDP_(PPP)_per_capita")
)
display(gdp_plot)
```

## Átlagfizetés alapú elemzés. 
A körök színei a kontinenseket jelölik, a körök méreteti az adott ország nepességére utal.

```python
circle_df_poor = generate_circle(x = 400, y=15, radius=16, ratio=35)
circle_df_rich = generate_circle(x = 3100, y=25, radius=17, ratio=90)
salary_plot = (
    ggplot(df_pivoted5, aes(x="salary", y="cancelled_to_converted_ratio (%)", size='sqrt_pop')) +
    geom_point(aes(color="continent"), alpha=0.5 ) + #color='#2544d8'
    xlim(-500, 6500) +
    scale_size_continuous(range=(2, 30)) +
    geom_smooth() +
    theme_minimal() +
    guides(size=False, shape=False, fill=False, alpha=False, linetype=False) +
    theme(axis_text_x=element_text(angle=45), figure_size=(9, 8), plot_caption=element_text(size=10, color="gray", ha="left"), legend_key_size=15, legend_text=element_text(size=14)) +
    geom_path(data=circle_df_poor, mapping=aes(x="x", y="y"), color="#2d3538", size=1, alpha =0.6) +
    geom_path(data=circle_df_rich, mapping=aes(x="x", y="y"), color="#2544d8", size=1, alpha =0.6) +
    geom_text(aes(label="country_name"), va="top", ha="center", size=8, angle=0) +
    labs(title="Országonkénti konverziós arány az egy főre eső \n                 jövedelem függvényében", x="Havi átlagos nettó fizetés [$]", y="Konverziós arány [%]") +
    annotate("text", x=800, y=2, label="← Fejlődő országok, alacsonyabb bevétellel", ha="left") +
    annotate("text", x=200, y=33, label="Fejlett államok →", ha="left") +
    labs(caption="Forrás: https://countrycassette.com/average-monthly-salary-by-country-2023/")
)
salary_plot.save("./images/salary_plot.png", dpi=300, transparent=True)
display(salary_plot)
```
